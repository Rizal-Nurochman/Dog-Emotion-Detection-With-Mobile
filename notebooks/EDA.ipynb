## Import Library and Variables

```python
from pathlib import Path
from collections import Counter
from PIL import Image
import matplotlib.pyplot as plt
import yaml

CONFIG  = yaml.safe_load(Path("configs/config.yaml").read_text())
RAW_DIR = Path(CONFIG["data"]["raw_dir"])
CLASSES = CONFIG["data"]["classes"]
print("kelas:", CLASSES)
```

## Distribusi Data Gambar

```python
counts = {cls: len(list((RAW_DIR / cls).glob("*"))) for cls in CLASSES}
print(counts)

plt.figure(figsize=(6, 4))
plt.bar(counts.keys(), counts.values())
plt.title("Distribusi Jumlah Gambar per Kelas")
plt.xlabel("Kelas")
plt.ylabel("Jumlah gambar")
plt.show()
```

```python
plt.figure(figsize=(12, 3))
for i, cls in enumerate(CLASSES):
    first_img = next((RAW_DIR / cls).glob("*"))
    img = Image.open(first_img)
    plt.subplot(1, len(CLASSES), i + 1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis("off")
plt.show()
```

```python
sizes = []
for cls in CLASSES:
  for img_path in list((RAW_DIR / cls).glob("*"))[:100]:
    with Image.open(img_path) as im:
      sizes.append(im.size)

widths  = [w for w, h in sizes]
heights = [h for w, h in sizes]
print(f"lebar : min={min(widths)}  max={max(widths)}")
print(f"tinggi: min={min(heights)} max={max(heights)}")
```
